```python
%%writefile app.py
# ===============================
# AI PHISHING EMAIL DETECTOR DEMO
# ===============================

import streamlit as st
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

# -------------------------------
# Dataset (Embedded)
# -------------------------------
data = {
    "text": [
        "Your account has been suspended. Click here to verify immediately.",
        "Dear customer, your bank account is on hold. Login now.",
        "Congratulations! You won a prize. Claim now.",
        "Please find attached the meeting minutes from today.",
        "Let us schedule a meeting for project discussion.",
        "Invoice attached for your recent purchase. Thank you.",
        "Urgent action required! Update your password now.",
        "Reminder about your appointment tomorrow."
    ],
    "label": [1, 1, 1, 0, 0, 0, 1, 0]
}

df = pd.DataFrame(data)

def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^a-z\s]", "", text)
    return text

df["text"] = df["text"].apply(clean_text)

model = Pipeline([
    ("tfidf", TfidfVectorizer(stop_words="english")),
    ("clf", LogisticRegression())
])

model.fit(df["text"], df["label"])

# -------------------------------
# Streamlit UI
# -------------------------------
st.set_page_config(page_title="Phishing Email Detector")

st.title("üîê AI-Powered Phishing Email Detector")
st.write("BSc Cyber Security Project Demo")

email = st.text_area("üìß Paste Email Content")

if st.button("Analyze Email"):
    if email.strip() == "":
        st.warning("Please enter an email.")
    else:
        cleaned = clean_text(email)
        prediction = model.predict([cleaned])[0]
        confidence = model.predict_proba([cleaned]).max() * 100

        if prediction == 1:
            st.error(f"üö® PHISHING EMAIL\nConfidence: {confidence:.2f}%")
        else:
            st.success(f"‚úÖ LEGITIMATE EMAIL\nConfidence: {confidence:.2f}%")
```
